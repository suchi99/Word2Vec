Dataset link: https://www.kaggle.com/datasets/datafiniti/hotel-reviews

```python
from gensim.models.word2vec import Word2Vec
```

```python
import re
from sklearn import feature_extraction 
stop_words = feature_extraction.text.ENGLISH_STOP_WORDS
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer

def preprocess(text):
    if isinstance(text, str):  # Check if text is a string
        text = text.lower()  # Lowercase the text
        text = re.sub(r'[^\w\s]', '', text)  # Remove punctuations
        text = re.sub(r'\d+', '', text)  # Remove numbers
        text = " ".join(text.split()) # Strip whitespaces
        text = text.split()  # Split into words
        text = [x for x in text if x not in stop_words]  # Remove stopwords
        # text = [x for x in text if x not in ["dr", "doctor"]]  # Remove task specific stopwords
        text = " ".join(text)
        # stemmer_ps = PorterStemmer()  
        # text = [stemmer_ps.stem(word) for word in text.split()]  # Stemming
        # text = " ".join(text)
        # lemmatizer = WordNetLemmatizer()
        # text = [lemmatizer.lemmatize(word) for word in text.split()]  # Lemmatization
        # text = " ".join(text)
        return text
    else:
        return ''  # Return an empty string for non-string inputs
```

```python
import pandas as pd
data = pd.read_csv(r"C:\Users\vavil\Downloads\archive (13)\Datafiniti_Hotel_Reviews.csv")
```

```python
data
```

```python
data['review_processed']=data['reviews.text'].apply(lambda x:preprocess(x))
data['review_processed']=data['review_processed'].apply(lambda x:x.split())
```

```python
data
```

```python
model = Word2Vec(sentences=data['review_processed'].tolist(), vector_size=100, sg=1,min_count=2,window=5,workers=50,seed=10,epochs=50)
#vector_size (int, optional) – Dimensionality of the word vectors.
#min_count (int, optional) – Ignores all words with total frequency lower than this.
#window (int, optional) – Maximum distance between the current and predicted word within a sentence.
#workers (int, optional) – Use these many worker threads to train the model (=faster training with multicore machines).
#sg ({0, 1}, optional) – Training algorithm: 1 for skip-gram; otherwise CBOW.
#epochs (int, optional) – Number of iterations (epochs) over the corpus. (Formerly: iter)
```
